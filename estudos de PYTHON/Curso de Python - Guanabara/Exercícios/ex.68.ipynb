```python
"""
Ex.68 Faça um programa que jogue par ou ímpar com o computador. O jogo só será interrompido quando o jogador perder, mostrando o total de vitórias consecutivas que ele conquistou no final do jogo.
"""
from random import randint
from colorama import Fore, Style
venceu = True
vitorias = 0
cont = 1

while True:
    escolha_jogador = ' '

    while escolha_jogador.lower() not in ['p', 'i', 'par', 'impar', 'ímpar']:
        escolha_jogador = input("Escolha entre 'Par' ou 'impar' [P/I]").strip().lower()
        if escolha_jogador not in ['p', 'i', 'par', 'impar', 'ímpar']:
            print('Escolha Par ou Ímpar [p / i].')

    print(f'\n --- {cont}ª rodada. ---')
    numero_jogador = int(input('Escolha um número entre 0 e 10.'))

    print(f'O jogador escolheu: {escolha_jogador}')
    print(f'O jogador escolheu: {numero_jogador}')
    numero_computador = random.randint(0, 10)
    print(f'O Computador escolheu: {numero_computador}')
    total = 0

    if escolha_jogador in ['p', 'par']:
        total = numero_jogador + numero_computador
        if total % 2 == 0:
            print(Fore.GREEN + f'O Jogador Ganhou, pois {total} é par.')
            print(Style.RESET_ALL + '-=' * 20)
            vitorias += 1
            cont += 1
        else:
            print(Fore.RED + f'O Computador Ganhou, pois {total} é ímpar.')
            print(Style.RESET_ALL + '-=' * 20)
            venceu = False  
            break

    elif escolha_jogador in ['i', 'impar', 'ímpar']:
        total = numero_jogador + numero_computador
        if total % 2 == 0:
            print(Fore.RED + f'O Computador Ganhou, pois {total} é par.')
            print(Style.RESET_ALL + '-=' * 20)
            venceu = False
            break       
        else:
            print(Fore.GREEN +f'O Jogador Ganhou, pois {total} é ímpar.')
            print(Style.RESET_ALL + '-=' * 20)
            vitorias += 1
            cont += 1

print(Fore.BLUE +f'\nO jogador conseguiu {vitorias} vitórias consecutivas.')
print(Fore.RED +'\nGAME OVER.')
print(Fore.YELLOW +'\nPrograma Encerrado!')

```
