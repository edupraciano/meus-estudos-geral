```python
"""
Ex.59 Crie um programa que leia dois valores e mostre um menu na tela:
[ 1 ] somar
[ 2 ] multiplicar
[ 3 ] maior
[ 4 ] novos números
[ 5 ] sair do programa
Seu programa deverá realizar a operação solicitada em cada caso.
"""

n1 = int(input('Digite o primeiro Número:'))
n2 = int(input('Digite o Segundo Número:'))
opção = ''
maior = 0
while opção != 5:
    print('O que Você deseja fazer com esses números?')
    print('''
    [ 1 ] somar
    [ 2 ] multiplicar
    [ 3 ] maior
    [ 4 ] novos números
    [ 5 ] sair do programa''')
    print('')
    opção = int(input('Escolha uma opção (1/2/3/4/5): '))

    if opção == 1:
        soma = n1 + n2
        print('Você decidiu SOMAR os números. O resultado é: {}'.format(soma))
        print('')
    elif opção == 2:
        mult = n1 * n2
        print('Você decidiu MULTIPLICAR os números. O resultado é: {}'.format(mult))
        print('')
    elif opção == 3:
        if n1 > n2:
            maior = n1
            print(
                'Você decidiu verificar qual dos dois números era o MAIOR. O resultado é: {}'.format(maior))
            print('')
        elif n1 < n2:
            maior = n2
            print(
                'Você decidiu verificar qual dos dois números era o MAIOR. O resultado é: {}'.format(maior))
            print('')
        else:
            print('Os números são iguais.')
            print('Você decidiu verificar qual dos dois números era o MAIOR. ENTRETANTO, isso não é possivel, pois os números são iguais')
            print('')
    elif opção == 4:
        print('Você decidiu entrar com NOVOS NÚMEROS.')
        print('')
        n1 = int(input('Digite o primeiro Número:'))
        n2 = int(input('Digite o Segundo Número:'))
    elif opção == 5:
        print('Fim do Menu ')
print('O programa foi encerrado.')
```
