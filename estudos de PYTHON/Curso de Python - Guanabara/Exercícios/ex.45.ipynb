```python
"""
Ex.45 Crie um programa que faça o computador jogar Jokenpô com você.
>> COMPUTADOR: Vamos jogar Pedra, Papel, Tesoura!
As regras são as seguintes:
- Papel vence Pedra e perde para Tesoura
- Pedra vence Tesoura e perde para Papel
- Tesoura vence Papel e perde para Pedra
"""
import random
from time import sleep

opt = ["pedra", "papel", "tesoura"]
escolhaDoPc = random.choice(opt)

escolhaDoJogador = (
    str(input('Escolha entre: "pedra", "papel" ou "tesoura".')).lower().strip()
)

if escolhaDoPc == "pedra" and escolhaDoJogador == "tesoura":
    print("JO")
    sleep(1)
    print("KEN")
    sleep(1)
    print("PÔ!!!")
    print(
        "Hahaha! Eu venci, pois {} que eu escolhi ganha de {} que você escolheu. Vamos tentar mais uma vez?".format(
            escolhaDoPc, escolhaDoJogador
        )
    )
elif escolhaDoPc == "papel" and escolhaDoJogador == "pedra":
    print("JO")
    sleep(1)
    print("KEN")
    sleep(1)
    print("PÔ!!!")
    print(
        "Hahaha! Eu venci, pois {} que eu escolhi ganha de {} que você escolheu. Vamos tentar mais uma vez?".format(
            escolhaDoPc, escolhaDoJogador
        )
    )
elif escolhaDoPc == "tesoura" and escolhaDoJogador == "papel":
    print("JO")
    sleep(1)
    print("KEN")
    sleep(1)
    print("PÔ!!!")
    print(
        "Hahaha! Eu venci, pois {} que eu escolhi ganha de {} que você escolheu. Vamos tentar mais uma vez?".format(
            escolhaDoPc, escolhaDoJogador
        )
    )
elif escolhaDoPc == escolhaDoJogador:
    print("JO")
    sleep(1)
    print("KEN")
    sleep(1)
    print("PÔ!!!")
    print(
        "Empatamos, pois nós dois escolhemos {}. Vamos tentar outra vez.".format(
            escolhaDoPc
        )
    )
elif escolhaDoJogador == "pedra" and escolhaDoPc == "tesoura":
    print("JO")
    sleep(1)
    print("KEN")
    sleep(1)
    print("PÔ!!!")
    print(
        "Puxa, dessa vez eu perdi, pois {} que você escolheu ganha de {} que eu escolhi. Vamos tentar mais uma vez?".format(
            escolhaDoJogador, escolhaDoPc
        )
    )
elif escolhaDoJogador == "papel" and escolhaDoPc == "pedra":
    print("JO")
    sleep(1)
    print("KEN")
    sleep(1)
    print("PÔ!!!")
    print(
        "Puxa, dessa vez eu perdi, pois {} que você escolheu ganha de {} que eu escolhi. \nVamos tentar mais uma vez?".format(
            escolhaDoJogador, escolhaDoPc
        )
    )
elif escolhaDoJogador == "tesoura" and escolhaDoPc == "papel":
    print("JO")
    sleep(1)
    print("KEN")
    sleep(1)
    print("PÔ!!!")
    print(
        "Puxa, dessa vez eu perdi, pois {} que você escolheu ganha de {} que eu escolhi. \nVamos tentar mais uma vez?".format(
            escolhaDoJogador, escolhaDoPc
        )
    )
```
