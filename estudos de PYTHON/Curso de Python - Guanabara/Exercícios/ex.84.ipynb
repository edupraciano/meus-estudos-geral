```python
"""
Ex.84 Faça um programa que leia nome e peso de várias pessoas, guardando tudo em uma lista. No final, mostre:
A) Quantas pessoas foram cadastradas. 
B) Uma listagem com as pessoas mais pesadas. 
C) Uma listagem com as pessoas mais leves.
"""
# A referencia que eu estabeleci pra quem é mais leve ou mais pesado são os seguintes pesos:
# 100 kg ou mais => pesados
# menos de 50 kg => leves

leves = []
pesadas = []
normal = []
totLeves = totPesadas = totNormal = 0
cont = 1

while True:
    nome = str(input(f"Digite o nome da {cont}ª pessoa: "))
    peso = float(input(f"Digite o peso da {cont}ª pessoa: "))

    if peso < 50:
        leves.append([nome, peso])
        print(f"Nome da {cont}ª pessoas leve: {nome}")
        print(f"Peso da {cont}ª pessoas leve: {peso} kg.")
        print("-=" * 20)
        totLeves += 1
    elif peso >= 100:
        pesadas.append([nome, peso])
        print(f"Nome da {cont}ª pessoas pesada: {nome}")
        print(f"Peso da {cont}ª pessoas pesada: {peso} kg.")
        print("-=" * 20)
        totPesadas += 1
    else:
        normal.append([nome, peso])
        print(f"Nome da {cont}ª pessoas pesada: {nome}")
        print(f"Peso da {cont}ª pessoas pesada: {peso} kg.")
        print("-=" * 20)
        totNormal += 1
    cont += 1
    sair = ""
    while sair not in ["S", "N"]:
        sair = str(input("Deseja SAIR? Digite [S/N]")).strip().upper()[0]
        if sair not in ["S", "N"]:
            print("entrada Inválida, Digite S ou N")
    if sair == "S":
        break

print(f"\nAo todo foram cadastradas {cont-1} pessoas.\n")

if pesadas:
    print(f"Foram cadastradas {totPesadas} pessoas com mais de 100 kg. São elas:")
    for pessoa in pesadas:
        print(f"Nome: {pessoa[0]}, Peso: {pessoa[1]} kg.")

if leves:
    print(f"\nForam cadastradas {totLeves} pessoas com menos de 50 kg. São elas:")
    for pessoa in leves:
        print(f"Nome: {pessoa[0]}, Peso: {pessoa[1]} kg.")

if normal:
    print(
        f"\nForam cadastradas {totNormal} pessoas com peso normal (acima de 50 kg e menos de 100Kg). São elas:"
    )
    for pessoa in normal:
        print(f"Nome: {pessoa[0]}, Peso: {pessoa[1]} kg.")
```
